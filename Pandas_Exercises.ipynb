### Write a Pandas program to filter those records where WHO region matches with multiple values (Africa, Eastern Mediterranean, Europe) from world alcohol consumption dataset

```python
df[df["WHO region"].isin(["Africa", "Eastern Mediterranean", "Europe"])]
```

### Write a Pandas program to filter those records which not appears in a given list from world alcohol consumption dataset.

```python
df[~df["WHO region"].isin(["Africa", "Eastern Mediterranean", "Europe"])]
```

### Write a Pandas program to filter all records where the average consumption of beverages per person from .5 to 2.50 in world alcohol consumption datase

```python
df[(df["Display Value"] < 2.5) & (df["Display Value"] > 0.5)]
```

### Write a Pandas program to find number of consumption of wine per person greater than 2 in world alcohol consumption dataset.

```python
df[(df["Beverage Types"] == "Wine") & (df["Display Value"] > 2)].count()
```

### Write a Pandas program to filter rows based on row numbers ended with 0, like 0, 10, 20, 30 from world alcohol consumption dataset.

```python
df.filter(regex='0$', axis=0)
```

### Create a dataframe of ten rows, four columns with random values. Convert some values to nan values. Write a Pandas program which will highlight the nan values.

```python
np.random.seed(300)
df = pd.DataFrame({'A': np.linspace(1, 10, 10)})
df = pd.concat([df, pd.DataFrame(np.random.randn(10, 4), columns=list('BCDE'))],
               axis=1)
df.iloc[0, 3] = np.nan
df.iloc[2, 2] = np.nan
df.iloc[0, 1] = np.nan
df.iloc[5, 2] = np.nan
df.iloc[5, 3] = np.nan
print(df)
df.style.highlight_null(null_color='red')
```

### Create a dataframe of ten rows, four columns with random values. Write a Pandas program to highlight dataframe’s specific columns with different colors.

```python
np.random.seed(1)
df = pd.DataFrame({'A': np.linspace(1, 10, 10)})
df = pd.concat([df, pd.DataFrame(np.random.randn(10, 4), columns=list('BCDE'))],
               axis=1)
print(df)

def highlight_cols(x):
    df = x.copy()
    df.loc[:,:] = 'background-color: red'
    df[['B','C','E']] = 'background-color: grey'
    return df    

df.style.apply(highlight_cols, axis=None)
```
